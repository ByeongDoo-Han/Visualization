```python
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid", palette='muted')

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
cor = iris.corr()
sns.heatmap(cor)
```

```python
iris = pd.melt(iris, "species", var_name='measurement')
iris
```

```python
sns.scatterplot(data=iris, x='measurement' , y='value')
```

```python
sns.swarmplot(data=iris, x='measurement' , y='value')
```

```python
tips = sns.load_dataset('tips')
tips.sample()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```
